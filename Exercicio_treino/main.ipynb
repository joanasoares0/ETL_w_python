```python
#imports

import pandas as pd
```

# Extract

```python
# read csv 
df = pd.read_csv('World_Happiness_Report_2024.csv', sep=',')
df.head()
```

# Transform

```python
column_renaming: dict = {
    "Country name": "pais",
    "year": "ano",
    "Life Ladder": "escada_da_vida",
    "Log GDP per capita": "log_pib_per_capita",
    "Social support": "suporte_social",
    "Healthy life expectancy at birth": "expectativa_de_vida_saudavel_ao_nascer",
    "Freedom to make life choices": "liberdade_de_escolhas_de_vida",
    "Generosity": "generosidade",
    "Perceptions of corruption": "perceção_de_corrupcao",
    "Positive affect": "afeto_positivo",
    "Negative affect": "afeto_negativo",
    }

df = df.rename(columns=column_renaming)
df.columns
```

```python
def classify_happiness(score):
    if score > 6.5:
        return 'Alta'
    elif score >= 5.0:
        return 'Média'
    else:
        return 'Baixa'
    
df['felicidade'] = df[
    ['escada_da_vida', 'log_pib_per_capita', 'suporte_social',
       'expectativa_de_vida_saudavel_ao_nascer',
       'liberdade_de_escolhas_de_vida', 'generosidade',
       'perceção_de_corrupcao', 'afeto_positivo', 'afeto_negativo']
    ].median(axis=1)
# or if we only had numeric vlaues to calc the mean:
# df['felicidade'] = df.select_dtypes(include='number').mean(axis=1)

df['nivel_felicidade'] = df['felicidade'].apply(classify_happiness)
df.groupby(['pais', 'nivel_felicidade'])['nivel_felicidade'].count()
```

```python
latin_america = [
    "Argentina",
    "Bolivia",
    "Brazil",
    "Chile",
    "Colombia",
    "Costa Rica",
    "Cuba",
    "Dominican Republic",
    "Ecuador",
    "El Salvador",
    "Guatemala",
    "Haiti",
    "Honduras",
    "Mexico",
    "Nicaragua",
    "Panama",
    "Paraguay",
    "Peru",
    "Uruguay",
    "Venezuela",
]

df_latin_america = df[ df['pais'].isin(latin_america) ].sort_values('pais')
df_latin_america
```

# Load

```python
df_latin_america.to_csv('relatorio_felicidade_latam_2024.csv', index=False)
```
