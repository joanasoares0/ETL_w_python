###Imagine esse arquivo exemplo.csv:
###id	idade
###1	   30
###2	   25
###3	   trinta
###4	   40

Essa coluna idade tem um problema: a terceira linha contém o texto "trinta" em vez de um número.

O que acontece ao ler isso com pandas.read_csv():
python
import pandas as pd

df = pd.read_csv("exemplo.csv")

O pandas, ao ler os primeiros pedaços do arquivo (chunks), pode achar que idade é numérica. Quando encontrar "trinta", ele fica em dúvida: será que essa coluna é int, float, ou str?

Por isso ele emite o DtypeWarning dizendo:

"Essa coluna tem tipos mistos. Melhor você revisar isso!"

Como resolver de forma simples:
Usando low_memory=False:

df = pd.read_csv("exemplo.csv", low_memory=False)

Isso faz o pandas ler o arquivo inteiro antes de decidir os tipos.

Ou forçando como texto:

df = pd.read_csv("exemplo.csv", dtype={'idade': str})
Agora ele trata todos os valores da coluna como texto, evitando confusão.

##Conclusão:
Esse tipo de erro é comum em bases com dados sujos ou misturados. Sempre que vir esse aviso, pense:

“Será que essa coluna deveria ter só números?”

“Posso forçar ela a ser string?”

“Preciso tratar esses dados antes?”

```python
df = pd.read_csv('movies_metadata.csv', sep=',', low_memory=False)
```

#df.head()
Mostra as 5 primeiras linhas do DataFrame.
É útil para entender a estrutura da tabela, os nomes das colunas e os primeiros dados.

```python
print(df.head())
```

#Selecionar colunas específicas
Seleciona somente as colunas que vamos usar na análise.
Isso ajuda a deixar o DataFrame mais leve e focado.

```python
colunas_utilizadas = ['title', 'budget', 'popularity', 'release_date', 'revenue']
df = df[colunas_utilizadas]
```

#df.dropna()
Remove as linhas que possuem valores nulos.
É um passo importante para evitar erros nas transformações seguintes.

```python
df = df.dropna()
```

#pd.to_numeric(..., errors='coerce')
Converte colunas de texto para número.
Se algum valor for inválido, ele vira NaN (nulo).
Aqui, usamos para garantir que 'budget', 'popularity' e 'revenue' sejam tratados como números.

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
```

Depois da conversão, removemos novamente linhas com valores nulos (por exemplo, aquelas que tinham texto em vez de número).

```python
df = df.dropna()
```

#pd.to_datetime(...)
Converte a coluna de datas, que está como texto, para o formato de data.
Isso permite fazer análises por ano, mês etc.

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

#Criar coluna com o ano
Agora que a data está formatada, podemos extrair só o ano da data de lançamento.

```python
df['release_year'] = df['release_date'].dt.year
```

#Converter valores em milhões
Divide os valores por 1 milhão para transformar o número em milhões., e arredondamos com .round(2) para deixar o número mais limpo.

```python
df['budget_milhoes'] = (df['budget'] / 1_000_000).round(2)
df['revenue_milhoes'] = (df['revenue'] / 1_000_000).round(2)
```

Mostrar dados transformados
Mostra uma amostra do DataFrame já transformado.

```python
print("\nDados transformados:")
print(df[['title', 'release_year', 'budget_milhoes', 'revenue_milhoes', 'popularity']].head())
```

#groupby()
Agrupa os dados por uma coluna (ex: ano de lançamento) e permite calcular agregações como:

.mean(): média

.sum(): soma

.count(): contagem

```python
media_orcamento = df.groupby('release_year')['budget_milhoes'].mean()
```

#to_csv()
Salva o DataFrame em um novo arquivo .csv.

index=False: não salva a coluna de índice (número das linhas).

```python
df.to_csv('filmes_transformados.csv', index=False)
```
